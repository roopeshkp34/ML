## Making figures with numpy array

```python
import numpy as np
```

```python
# Create some data
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# plot the data and create a line plot
fix, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# use the same data to make a scatter
fix, ax = plt.subplots()
ax.scatter(x, np.exp(x));
```

```python
# another scatter plot
fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# make a plot from dict
nut_butter_prices = {"Almond_butter": 10, "peanut_butter": 8, "cashew_butter": 12}
fig, ax = plt.subplots()
ax.bar(nut_butter_prices.keys(),nut_butter_prices.values())
ax.set(title="Nut butter store",
       ylabel="Price");
```

# Two options for subplots

```python
# subplo toption 1
fix, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, ncols=2, figsize=(10,5))

# plot to each different axis
ax1.plot(x, x/2);
ax2.plot(np.random.random(10), np.random.random(10));
ax3.bar(nut_butter_prices.keys(), nut_butter_prices.values());
ax4.hist(np.random.randn(1000));
```
